```python
import matplotlib.pyplot as plt

import pandas as pd

new_data = {'category': ['AB212', 'AC212', 'AZ887', 'DB001', 'ET683', 'DU054', 'OP689', 'GH518', 'NMO', 'EE700', 'IL826', 'MK553'],
 'price': [234, 110, 550, 40, 1180, 150, 302, 480, 750,23,400,200]
}

```

```python
df=pd.DataFrame.from_dict (new_data)
df
```

```python
fig = plt.figure(figsize = (9,6))
ax1 = plt.subplot (111) 
ax1.barh(df [ 'category'], df [ 'price'])

plt.show()
```

```python
round(df.price.mean())
```

```python
df.append({ 'category':'mean','price': round(df.price.mean())}, ignore_index=True)
```

```python
fig = plt.figure(figsize = (9,6))

ax1 = plt.subplot (111)

#df = df.sort_values('price')

df.sort_values('price', inplace=True) 
ax1.barh(df['category' ], df ['price'])

plt.show()
```

```python
fig = plt.figure(figsize = (9,6))
ax1 = plt.subplot (111)


df.sort_values('price', inplace=True) #df = df.sort_values('price')

#setting different bar colr

ret = ax1.barh(df ['category'], df['price'], color = '#99ccff')

ret [7].set_color('#404040')

ret [10].set_color('#CC0000')

ax1.xaxis.grid(linestyle='--', linewidth ='0.5')

plt.show()
```
